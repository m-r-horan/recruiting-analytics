## Applications

```python
# ── MATCH CANDIDATES TO JOBS ──────────────────────────────────────

def get_target_level(years_exp):
    """
    Given a candidate's years of experience, return the job level
    they are most likely to apply for — with some realistic noise
    (people apply above and below their experience level)
    """
    if years_exp <= 2:
        # entry level — mostly applies to IC, occasionally stretches to Senior IC
        return random.choices(['IC', 'Senior IC'], weights=[0.85, 0.15], k=1)[0]
    elif years_exp <= 5:
        # mid level — could go IC or Senior IC
        return random.choices(['IC', 'Senior IC'], weights=[0.40, 0.60], k=1)[0]
    elif years_exp <= 9:
        # senior — mostly Senior IC, sometimes reaches for Manager
        return random.choices(['Senior IC', 'Manager'], weights=[0.70, 0.30], k=1)[0]
    elif years_exp <= 14:
        # experienced — Manager or Director
        return random.choices(['Manager', 'Director'], weights=[0.75, 0.25], k=1)[0]
    else:
        # very experienced — split between Manager and Director
        return random.choices(['Manager', 'Director'], weights=[0.50, 0.50], k=1)[0]


# build a lookup of jobs by level for efficient filtering
# e.g. jobs_by_level['IC'] = list of all IC job_ids
jobs_by_level = df_jobs.groupby('level')['job_id'].apply(list).to_dict()

# sources live on the application, not the candidate
# weighted — LinkedIn heaviest, recruiter outreach smallest
sources = ['LinkedIn', 'Company Website', 'Referral', 'Indeed', 'Recruiter Outreach']
source_weights = [0.42, 0.22, 0.16, 0.12, 0.08]

applications = []
app_id = 1  # running application counter

for _, candidate in df_candidates.iterrows():
    
    # determine how many jobs this candidate applies to
    # 80% apply to 1 job, 20% apply to 2 or 3
    if random.random() < 0.80:
        num_applications = 1
    else:
        num_applications = random.choices([2, 3], weights=[0.70, 0.30], k=1)[0]

    # get the job level this candidate is targeting based on experience
    target_level = get_target_level(candidate['years_experience'])

    # get list of jobs at that level — fall back to IC if none available
    available_jobs = jobs_by_level.get(target_level, jobs_by_level.get('IC', []))

    # pick random jobs from available pool without replacement
    # (a candidate shouldn't apply to the same job twice)
    num_applications = min(num_applications, len(available_jobs))
    selected_jobs = random.sample(available_jobs, num_applications)

    for job_id in selected_jobs:
        applications.append({
            'application_id': f'APP-{app_id:06d}',  # e.g. APP-000001
            'candidate_id':   candidate['candidate_id'],
            'job_id':         job_id,
            'source':         random.choices(sources, weights=source_weights, k=1)[0],
        })
        app_id += 1

df_applications_raw = pd.DataFrame(applications)

# sanity checks
print(f"total applications: {len(df_applications_raw)}")
print(f"avg applications per job: {len(df_applications_raw) / len(df_jobs):.1f}")
print(f"avg applications per candidate: {len(df_applications_raw) / len(df_candidates):.1f}")
print(df_applications_raw['source'].value_counts())

df_applications_raw.head()
```

```python
# ── GENERATE APPLICATIONS WITH FUNNEL ─────────────────────────────

# base pass rates at each funnel stage
base_rates = {
    'Applied':         0.20,  # resume screen — most filtered here
    'Phone Screen':    0.55,  # phone screen → interview
    'Interview':       0.50,  # interview → final round
    'Final Round':     0.55,  # final round → offer
    'Offer Extended':  0.80,  # offer acceptance rate
}

# source modifiers — referrals and recruiter outreach are pre-vetted
source_modifiers = {
    'Referral':           0.15,
    'Recruiter Outreach': 0.10,
    'LinkedIn':           0.00,
    'Company Website':    0.00,
    'Indeed':             0.00,
}

# experience fit sweet spots by job level
experience_ranges = {
    'IC':        (0,  5),
    'Senior IC': (4,  9),
    'Manager':   (7,  13),
    'Director':  (12, 25),
}

# time between stages in days by department
# engineering takes longest due to complex interview process
time_between_stages = {
    'Engineering':        {'Phone Screen': (5, 10), 'Interview': (10, 20), 'Final Round': (7, 14), 'Offer Extended': (3, 7)},
    'Product':            {'Phone Screen': (4, 8),  'Interview': (7, 14),  'Final Round': (5, 10), 'Offer Extended': (2, 5)},
    'Data & Analytics':   {'Phone Screen': (4, 8),  'Interview': (7, 14),  'Final Round': (5, 10), 'Offer Extended': (2, 5)},
    'Sales':              {'Phone Screen': (2, 5),  'Interview': (3, 7),   'Final Round': (3, 7),  'Offer Extended': (1, 3)},
    'Customer Success':   {'Phone Screen': (3, 6),  'Interview': (5, 10),  'Final Round': (4, 8),  'Offer Extended': (2, 4)},
    'Marketing':          {'Phone Screen': (3, 6),  'Interview': (5, 10),  'Final Round': (4, 8),  'Offer Extended': (2, 4)},
    'HR':                 {'Phone Screen': (3, 6),  'Interview': (5, 10),  'Final Round': (4, 8),  'Offer Extended': (2, 4)},
    'Finance':            {'Phone Screen': (3, 6),  'Interview': (5, 10),  'Final Round': (4, 8),  'Offer Extended': (2, 4)},
    'Operations':         {'Phone Screen': (3, 6),  'Interview': (5, 10),  'Final Round': (4, 8),  'Offer Extended': (2, 4)},
    'Legal & Compliance': {'Phone Screen': (3, 6),  'Interview': (5, 10),  'Final Round': (4, 8),  'Offer Extended': (2, 4)},
}

# rejection reasons by stage
rejection_reasons = {
    'Applied':        ['Resume did not meet requirements', 'Insufficient experience', 'Position filled'],
    'Phone Screen':   ['Failed phone screen', 'Salary expectations misaligned', 'Communication concerns'],
    'Interview':      ['Technical skills gap', 'Culture fit concerns', 'Stronger candidate selected'],
    'Final Round':    ['Stronger candidate selected', 'Leadership concerns', 'Culture fit concerns'],
    'Offer Extended': ['Candidate declined offer', 'Candidate accepted another offer'],
}

def get_experience_modifier(years_exp, job_level):
    """modifier based on how well candidate experience matches job level"""
    low, high = experience_ranges[job_level]
    if years_exp < low:
        return -0.20   # underqualified
    elif years_exp > high:
        return -0.10   # overqualified
    else:
        return 0.00    # good fit

def get_pass_rate(stage, source, years_exp, job_level):
    """calculate probability of advancing past a stage"""
    base    = base_rates[stage]
    src_mod = source_modifiers.get(source, 0)
    exp_mod = get_experience_modifier(years_exp, job_level)
    return max(0.0, min(1.0, base + src_mod + exp_mod))

def get_withdraw_prob(stage):
    """probability of withdrawing at each stage — increases later in funnel"""
    withdraw_probs = {
        'Applied':        0.01,
        'Phone Screen':   0.03,
        'Interview':      0.05,
        'Final Round':    0.08,
        'Offer Extended': 0.00,  # handled by offer acceptance rate
    }
    return withdraw_probs[stage]

# ── MAIN APPLICATION GENERATION LOOP ──────────────────────────────

# build lookups for efficient access during loop
candidate_lookup = df_candidates.set_index('candidate_id').to_dict('index')
job_lookup       = df_jobs.set_index('job_id').to_dict('index')
apps_by_job      = df_applications_raw.groupby('job_id')
stages           = ['Applied', 'Phone Screen', 'Interview', 'Final Round', 'Offer Extended']

final_applications = []

for job_id, job_apps in apps_by_job:
    job        = job_lookup[job_id]
    dept       = job['department']
    job_level  = job['level']
    job_status = job['status']
    open_date  = job['open_date']

    # track whether this job has been filled — enforces 1 hire per job
    job_filled = False

    # shuffle so hiring order is random, not based on application order
    applicants = job_apps.to_dict('records')
    random.shuffle(applicants)

    for app in applicants:
        candidate_id = app['candidate_id']
        source       = app['source']
        candidate    = candidate_lookup[candidate_id]
        years_exp    = candidate['years_experience']

        # initialise all fields as None
        current_stage    = 'Applied'
        outcome          = 'Active'
        rejection_reason = None
        applied_date     = None
        screen_date      = None
        interview_date   = None
        final_round_date = None
        offer_date       = None
        hire_date        = None

        # applied date — random date after job opened
        applied_date = open_date + timedelta(days=random.randint(1, 30))
        current_date = applied_date

        # ── job already filled — reject at Applied ─────────────────────
        if job_filled:
            current_stage    = 'Applied'
            outcome          = 'Rejected'
            rejection_reason = 'Position filled'

        # ── on hold — freeze everyone at Applied, no outcome ───────────
        elif job_status == 'On Hold':
            current_stage = 'Applied'
            outcome       = 'Active'

        # ── open and closed jobs — run through funnel ──────────────────
        else:
            for stage in stages:
                current_stage = stage

                # check withdrawal at each stage
                if random.random() < get_withdraw_prob(stage):
                    outcome = 'Withdrew'
                    break

                pass_rate = get_pass_rate(stage, source, years_exp, job_level)

                if stage == 'Offer Extended':
                    if job_status == 'Open':
                        # open jobs — candidate reached offer but no decision yet
                        # keep them Active at Offer Extended stage
                        outcome       = 'Active'
                        current_stage = 'Offer Extended'
                        offer_date    = current_date
                        break

                    # closed jobs only — make the hire decision
                    if random.random() < pass_rate:
                        # candidate accepted offer — hired
                        outcome       = 'Hired'
                        current_stage = 'Hired'
                        offer_date    = current_date
                        hire_date     = current_date + timedelta(
                            days=random.randint(*time_between_stages[dept]['Offer Extended'])
                        )
                        job_filled = True
                    else:
                        # candidate declined offer
                        outcome          = 'Rejected'
                        current_stage    = 'Offer Extended'
                        rejection_reason = 'Candidate declined offer'
                        offer_date       = current_date
                    break

                elif random.random() < pass_rate:
                    # candidate advances to next stage — calculate next stage date
                    next_stage_days = time_between_stages[dept].get(
                        stages[stages.index(stage) + 1]
                    )
                    if next_stage_days:
                        current_date = current_date + timedelta(
                            days=random.randint(*next_stage_days)
                        )

                    # record date for the stage just passed
                    if stage == 'Applied':
                        screen_date = current_date
                    elif stage == 'Phone Screen':
                        interview_date = current_date
                    elif stage == 'Interview':
                        final_round_date = current_date
                    elif stage == 'Final Round':
                        offer_date = current_date

                else:
                    # candidate failed this stage — rejected
                    outcome          = 'Rejected'
                    rejection_reason = random.choice(rejection_reasons[stage])
                    break

            # closed jobs — if still active at end of funnel, reject
            if job_status == 'Closed' and outcome == 'Active':
                outcome          = 'Rejected'
                rejection_reason = 'Position filled'

            # open jobs — if still in funnel, keep as Active
            if job_status == 'Open' and outcome not in ['Hired', 'Rejected', 'Withdrew']:
                outcome = 'Active'

        final_applications.append({
            'application_id':   app['application_id'],
            'candidate_id':     candidate_id,
            'job_id':           job_id,
            'source':           source,
            'current_stage':    current_stage,
            'outcome':          outcome,
            'rejection_reason': rejection_reason,
            'applied_date':     applied_date,
            'screen_date':      screen_date,
            'interview_date':   interview_date,
            'final_round_date': final_round_date,
            'offer_date':       offer_date,
            'hire_date':        hire_date,
        })

df_applications = pd.DataFrame(final_applications)

# ── SANITY CHECKS ─────────────────────────────────────────────────
print(f"total applications: {len(df_applications)}")
print(f"\noutcome distribution:\n{df_applications['outcome'].value_counts()}")
print(f"\ncurrent stage distribution:\n{df_applications['current_stage'].value_counts()}")
print(f"\nsource distribution:\n{df_applications['source'].value_counts()}")

# verify max 1 hire per job
hires_per_job = df_applications[df_applications['outcome'] == 'Hired'].groupby('job_id').size()
print(f"\nmax hires per job: {hires_per_job.max()}")
print(f"jobs with at least 1 hire: {len(hires_per_job)}")
print(f"closed jobs: {len(df_jobs[df_jobs['status'] == 'Closed'])}")

# verify hires only on closed jobs
all_hires = df_applications[df_applications['outcome'] == 'Hired'].merge(
    df_jobs[['job_id', 'status']], on='job_id'
)
print(f"\nhires by job status:\n{all_hires['status'].value_counts()}")

df_applications.head()
```

```python
# ── ADD MESSINESS ─────────────────────────────────────────────────

# 1. blank ~15% of rejection reasons — recruiters forget to fill these in
rejected_mask = df_applications['outcome'] == 'Rejected'
df_applications.loc[rejected_mask & (np.random.random(len(df_applications)) < 0.15), 'rejection_reason'] = None

# 2. inconsistent source labels — different recruiters log differently
source_variants = {
    'LinkedIn':           ['LinkedIn', 'linkedin', 'LinkedIn Jobs', 'Linkedin'],
    'Indeed':             ['Indeed', 'indeed', 'Indeed.com'],
    'Company Website':    ['Company Website', 'company website', 'Career Page'],
    'Referral':           ['Referral', 'referral', 'Employee Referral'],
    'Recruiter Outreach': ['Recruiter Outreach', 'Outreach', 'Sourced'],
}
df_applications['source'] = df_applications['source'].apply(
    lambda s: random.choice(source_variants[s]) if random.random() < 0.25 else s
)

# 3. null out ~8% of screen dates and ~5% of interview dates
df_applications.loc[df_applications['screen_date'].notna() & 
    (np.random.random(len(df_applications)) < 0.08), 'screen_date'] = None
df_applications.loc[df_applications['interview_date'].notna() & 
    (np.random.random(len(df_applications)) < 0.05), 'interview_date'] = None

print("Messiness added ✅")
print(f"Missing rejection reasons: {df_applications['rejection_reason'].isna().sum()}")
print(f"Source variants: {df_applications['source'].value_counts()}")
```
